```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

if not os.path.exists('visualizations'):
    os.makedirs('visualizations')

data = pd.read_csv('titanic_cleaned.csv')

# 1. Line Plot: Age Distribution
plt.figure(figsize=(10, 6))
data['Age'].value_counts().sort_index().plot(kind='line')
plt.title('Age Distribution')
plt.savefig('visualizations/line_plot.png')
plt.close()

# 2. Scatter Plot: Age vs Fare
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', data=data)
plt.title('Age vs Fare')
plt.savefig('visualizations/scatter_plot.png')
plt.close()

# 3. Histogram: Passenger Class Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Pclass'], bins=3)
plt.title('Passenger Class Distribution')
plt.savefig('visualizations/histogram.png')
plt.close()

# 4. Bar Chart: Survival Rates
plt.figure(figsize=(10, 6))
sns.barplot(x='Sex', y='Survived', data=data)
plt.title('Survival Rates by Gender')
plt.savefig('visualizations/bar_chart.png')
plt.close()

# 5. Box Plot: Fare by Class
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=data)
plt.title('Fare by Class')
plt.savefig('visualizations/box_plot.png')
plt.close()

# 6. Violin Plot: Age by Gender
plt.figure(figsize=(10, 6))
sns.violinplot(x='Sex', y='Age', data=data)
plt.title('Age by Gender')
plt.savefig('visualizations/violin_plot.png')
plt.close()

# 7. Heatmap: Correlation
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.savefig('visualizations/heatmap.png')
plt.close()

# 8. Pair Plot: Numerical Features
pair_plot = sns.pairplot(data[['Age', 'Fare', 'Pclass', 'Survived']])
pair_plot.savefig('visualizations/pair_plot.png')
plt.close()
```
